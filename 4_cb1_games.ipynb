## Предобработка данных

```python
# делаем копию датасета, чтобы вносить в неё изменения
games = games1.copy()
```

```python
# переведем названия всех признаков в нижний регистр через нижнее подчеркивание (по PEP8)
games.columns = [c.replace(" ", "_").lower() for c in games.columns]
print(games.columns)
```

```python
# Выведем процент пропущенных значений для каждого столбца датафрейма
pd.DataFrame(games.isna().mean()).sort_values(by=0, ascending=False).style.format('{:.2%}').background_gradient('coolwarm')
```

```python
# посмотрим количество пропущенных значений в каждой колонке
games.isnull().sum()
```

```python
# проверка на дубликаты
games.duplicated().sum()
```

__Размышления о потенциальных причинах пропусков:__


- Critic_Score — оценка критиков (максимум 100) пропусков 51%, видимо не все заполняют оценку.
- Rating — рейтинг от организации ESRB 40% пропусков, не все игры классифицируют по рейтингу.
- User_Score — оценка пользователей (максимум 10), пользователи не обязаны заполнять оценку, 40% пользователей не оценивают игру.
- Year_of_Release — год выпуска, небольшое значение пропусков менее 10%, думаю эти строки можно удалить без потери качества данных.
- Отсутствующие значения в Name — название игры, Genre — жанр игры всего 2 строки, можно удалить также как и Year_of_Release, без потерь в качестве данных.

```python
games['year_of_release'] = games['year_of_release'].astype('Int64')
```

```python
# отсортируем значения в столбце user_score и выведем уникальные значения столбца
games['user_score'].sort_values().unique()
```

```python
# отсортируем значения в столбце critic_score и выведем уникальные значения столбца
games['critic_score'].sort_values().unique()
```

```python
# отсортируем значения в столбце rating и выведем уникальные значения столбца
games['rating'].sort_values().unique()
```

```python
games['user_score'] = games['user_score'].replace('tbd', np.NaN)

games['user_score'] = games['user_score'].astype(float) 
```

tbd отсутствие оценки заменим на nan

### Зависимость продаж от платформ. 

Две лидирующие платформы по продажам игр DS и PS2. Построим распределение продаж на этих платформах по годам.

```python
games_platform = games.pivot_table(index='platform', values='sum_sales',aggfunc='count')\
                .sort_values(by='sum_sales', ascending= True)

games_platform.plot(kind='bar', grid=True, figsize=(12,5)) 
plt.title('Зависимость продаж игр от платформ')
plt.xlabel('Платформа')
plt.ylabel('Частота продаж игр')
plt.xticks(rotation=45)
plt.show()
```

```python
#платформы с максимальными продажами
games_platform.loc[games_platform['sum_sales'] >= 974]
```

```python
pc = games.pivot_table(index=['platform', 'year_of_release'], values='sum_sales',aggfunc='count')\
                  .query('platform == ["PC"]').sort_values(by='year_of_release', ascending= True)

pc.plot(kind='bar', grid=True, figsize=(12,5),color="orange" ) 
plt.title('Зависимость продаж игр от платформ')
plt.xlabel('Платформа PC')
plt.ylabel('Частота продаж игр')
plt.xticks(rotation=45)
plt.show()
```

## Проверка гипотез


### Cредние пользовательские рейтинги платформ Xbox One и PC одинаковые;

* *Нулевая* гипотеза (**H<sub>0</sub>**) заключается в том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* *Альтернативная* гипотеза (**H<sub>1</sub>**) заключается в том, что средние пользовательские рейтинги платформ Xbox One и PC разные;


У нас две независимые выборки о рейтингах пользователей платформ Xbox One и PC, поэтому возьмем метод st.ttest_ind

```python
games_user_score_xone = games_filter.loc[games_filter['platform'] == 'XOne', 'user_score'].reset_index()
games_user_score_xone = games_user_score_xone.dropna()
print('Среднее значение рейтинга платформы XOne', games_user_score_xone['user_score'].mean())
```

```python
games_user_score_pc = games_filter.loc[games_filter['platform'] == 'PC','user_score'].reset_index()
games_user_score_pc = games_user_score_pc.dropna()
print('Среднее значение рейтинга платформы PC', games_user_score_pc['user_score'].mean())
```

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(games_user_score_xone['user_score'], games_user_score_pc['user_score'])

print('p-значение:', results.pvalue)


if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Не отвергаем нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. 
При уровне статистической значимости 5%.

### Cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые;

* *Нулевая* гипотеза (**H<sub>0</sub>**) заключается в том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые;
* *Альтернативная* гипотеза (**H<sub>1</sub>**) заключается в том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


У нас две независимые выборки о рейтингах жанров платформ Action и Sports, поэтому возьмем метод st.ttest_ind

```python
games_user_score_action = games_filter.loc[games_filter['genre'] == 'Action', 'user_score'].reset_index()

games_user_score_action = games_user_score_action.dropna()
print('Среднее значение рейтинга action', games_user_score_action['user_score'].mean())
```

```python
games_user_score_sports = games_filter.loc[games_filter['genre'] == 'Sports', 'user_score'].reset_index()
games_user_score_sports = games_user_score_sports.dropna()
print('Среднее значение рейтинга sports', games_user_score_sports['user_score'].mean())
```

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(games_user_score_action['user_score'], games_user_score_sports['user_score'])

print('p-значение:', results.pvalue)


if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```
